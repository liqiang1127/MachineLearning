# matplotlib基础

```python
import matplotlib as mpl
import matplotlib.pyplot as plt 
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,100)
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.plot(x, y)
plt.show()
```

```python
cosy = np.cos(x)
cosy.shape
```

```python
siny = y
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red")
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red",linestyle="--")
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red",linestyle="--")
plt.xlim(-5, 15)
plt.ylim(0,1.5)
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red",linestyle="--")
plt.axis([-1, 11, -2, 2])
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red",linestyle="--")
plt.xlabel("x axis")
plt.ylabel("y value")
plt.show()
```

```python
plt.plot(x, siny, label = "sin(x)")
plt.plot(x, cosy, color="red",linestyle="--", label="cos(x)")
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend()
plt.show()
```

```python
plt.plot(x, siny, label = "sin(x)")
plt.plot(x, cosy, color="red",linestyle="--", label="cos(x)")
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend()
plt.title("Welcome to the ML world")
plt.show()
```

## 散点图

```python
plt.scatter(x , siny)
plt.show()
```

```python
plt.scatter(x, siny)
plt.scatter(x, cosy, color="red")
plt.show()
```

```python
x = np.random.normal(0, 1, 10000)
y = np.random.normal(0, 1, 10000)

plt.scatter(x, y)
plt.show()
```

```python
x = np.random.normal(0, 1, 10000)
y = np.random.normal(0, 1, 10000)

plt.scatter(x, y, alpha=0.1)
plt.show()
```

```python

x = np.array([0,12])
y = np.array([12,0])
x1 = np.array([2,4])
y1 = np.array([4,2])

x2 = np.array([8, 9])
y2 = np.array([9, 8])

plt.scatter(x1, y1, label="A")
plt.scatter(x2, y2, label="B")
plt.scatter(3, 4, color="red")
plt.plot(x, y, label="svm")
plt.axis([0,12,0,12])
plt.legend()
plt.xlabel("A attribute")
plt.ylabel("B attribute")
plt.show()
```
